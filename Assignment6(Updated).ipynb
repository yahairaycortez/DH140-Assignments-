```python
#url https://www.gutenberg.org/files/1112/1112.txt
import nltk
import requests
from bs4 import BeautifulSoup
response = requests.get('https://www.gutenberg.org/files/1112/1112.txt')
```

```python
response
```

```python
_string = response.text
_string
```

```python
from nltk.tokenize import word_tokenize, sent_tokenize
```

```python
sent = sent_tokenize(_string)
print(sent)
```

```python
words = []
for s in sent:
    for w in word_tokenize(s):
        words.append(w)
```

```python
print(words)
```

```python
from nltk.corpus import stopwords
from string import punctuation
nltk.download('stopwords')
```

```python
print(punctuation)
```

```python
myStopWords = list(punctuation) + stopwords.words('english')

wordsNoStop = []
for i in words:
    if i not in myStopWords:
        wordsNoStop.append(i)
#print(words)
print(wordsNoStop)
```

```python
from nltk.collocations import *
from nltk.probability import FreqDist
freq = FreqDist(wordsNoStop)

for i in sorted(freq, key=freq.get, reverse=True)[:20]:
    print(i,freq[i])
```

```python
import pandas as pd
from pathlib import Path  
import glob

from nltk.tokenize import sent_tokenize, word_tokenize
from nltk.corpus import stopwords
from string import punctuation
```

```python
extrastop = ['``',"''","'re","'s","'re",'``',"''","'ll","--","\'\'","...",
             "n\'t",'one','would','use','subject','from',"\'m","\'ve"]
```

```python
myStopWords = list(punctuation) + stopwords.words('english') + extrastop
```

```python
[w for w in word_tokenize(movie_list[0].lower()) if w not in myStopWords]
```

```python
listOfNewsWords = []
for i in movie_list:
    listOfNewsWords.append([w for w in word_tokenize(i.lower()) if w not in myStopWords])
```

```python
listOfNewsWords[0]
```

```python
from nltk.stem.porter import PorterStemmer
p_stemmer = PorterStemmer()
```

```python
listOfStemmedWords = []
for i in listOfNewsWords:
    listOfStemmedWords.append([p_stemmer.stem(w) for w in i])
```

```python
listOfStemmedWords[0]
```

```python
!pip install gensim

from gensim import corpora, models
import gensim
```

```python
dictionary = corpora.Dictionary(listOfStemmedWords)
corpus = [dictionary.doc2bow(text) for text in listOfStemmedWords]
```

```python
ldamodel = gensim.models.ldamodel.LdaModel(corpus, 
                                           num_topics=2, 
                                           id2word = dictionary, 
                                           passes=20)
```

```python
for i in ldamodel.print_topics(num_topics=2, num_words=20):
    print(i)
    
#words such as film, plot, act, made seem to be good for describing the 2, but 'pitch', 'get', and other words
#don't seem to be good words to represent the topics
```
